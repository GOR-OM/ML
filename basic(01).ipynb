```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable (feature)
y = np.array([2, 4, 5, 4, 5])              # Dependent variable (target)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions using the model
y_pred = model.predict(X)

# Visualize the data and regression line
plt.scatter(X, y, label='Data points')
plt.plot(X, y_pred, color='red', label='Linear regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# Get the coefficients of the linear regression equation
slope = model.coef_[0]
intercept = model.intercept_

# Print the equation of the line
print(f"Linear Equation: y = {slope}x + {intercept}")
```

# Accuracy 

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(Y, model.predict(X))
print("Mean Squared Error:", mse)
```

```python
# Calculate R-squared (R2)
r2 = r2_score(Y, model.predict(X))
print("R-squared (R2):", r2)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Given data
X = np.array([1, 3]).reshape(-1, 1)
Y = np.array([2, 4])
model = LinearRegression()
model.fit(X, Y)

# Values of C
C = np.array([2, 3, 4, 5, 6, 7]).reshape(-1, 1)

# Predict Y values for C
y_pred = model.predict(C)

# Create a scatter plot of actual vs. predicted values
plt.scatter(Y, model.predict(X), label="Actual vs. Predicted")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")

# Add a 45-degree reference line (ideal prediction)
plt.plot([min(Y), max(Y)], [min(Y), max(Y)], color='red', linestyle='--', linewidth=2, label="Ideal Prediction")

plt.legend()
plt.title("Actual vs. Predicted Values")
plt.show()
```
